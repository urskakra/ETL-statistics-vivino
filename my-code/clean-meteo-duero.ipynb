```python
# import libraries

import pandas as pd
import numpy as np
import re
```

```python
# import dataset

duero_meteo = pd.read_csv('../data/Arandadeduero_RibDuero_meteo.csv')
```

```python
duero_meteo
```

```python
# drop first two rows

duero_meteo.drop(duero_meteo.index[0:2], inplace=True)
duero_meteo
```

```python
duero_meteo = duero_meteo.reset_index()
duero_meteo.info()
```

```python
# drop columns

duero_meteo.drop(columns = ['index', 'utc_offset_seconds', 'timezone', 'timezone_abbreviation'], inplace=True)
duero_meteo
```

```python
# rename columns

duero_meteo.rename(columns={'latitude': 'date', 'longitude': 'temp_duero', 'elevation': 'precip_duero'}, inplace=True)
duero_meteo
```

```python
# date: change type into string
# temp, precipitation: change type into float

duero_meteo.date = duero_meteo.date.apply(lambda x: str(x))
duero_meteo.temp_duero = duero_meteo.temp_duero.apply(lambda x: float(x))
duero_meteo.precip_duero = duero_meteo.precip_duero.apply(lambda x: float(x))

duero_meteo
```

```python
duero_meteo['date'].astype('string')
```

```python
duero_meteo.info()
```

```python
# group data by month (regex extract characters in date)
# mean temperature
# sum precipitation

# join on 'data' in sql, adding all 3 regions
```

```python
duero_meteo['date'] = duero_meteo['date'].str.extract('(.{7})')
duero_meteo
```

```python
# mean temperature, grouped by month

duero_temp = duero_meteo.groupby(by="date").temp_duero.mean()
duero_temp
```

```python
duero_precip = duero_meteo.groupby(by='date').precip_duero.sum()
duero_precip
```

```python
rib_duero = pd.DataFrame([duero_temp, duero_precip]).T
rib_duero
```

```python
# save clean dataset to rioja_clean.csv

rib_duero.to_csv("../data/duero_clean.csv", index=False)
```
