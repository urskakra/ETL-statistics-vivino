```python
# import libraries

import pandas as pd
import numpy as np
import re
```

```python
# import dataset

catalunya_meteo = pd.read_csv('../data/CorberaDEbre_Cataluna_meteo.csv')
```

```python
catalunya_meteo
```

```python
# drop first two rows

catalunya_meteo.drop(catalunya_meteo.index[0:2], inplace=True)
catalunya_meteo
```

```python
catalunya_meteo = catalunya_meteo.reset_index()
catalunya_meteo.info()
```

```python
# drop columns

catalunya_meteo.drop(columns = ['index', 'utc_offset_seconds', 'timezone', 'timezone_abbreviation'], inplace=True)
catalunya_meteo
```

```python
# rename columns

catalunya_meteo.rename(columns={'latitude': 'date', 'longitude': 'temp_catalunya', 'elevation': 'precip_catalunya'}, inplace=True)
catalunya_meteo
```

```python
# date: change type into string
# temp, precipitation: change type into float

catalunya_meteo.date = catalunya_meteo.date.apply(lambda x: str(x))
catalunya_meteo.temp_catalunya = catalunya_meteo.temp_catalunya.apply(lambda x: float(x))
catalunya_meteo.precip_catalunya = catalunya_meteo.precip_catalunya.apply(lambda x: float(x))

catalunya_meteo
```

```python
catalunya_meteo['date'].astype('string')
```

```python
catalunya_meteo.info()
```

```python
# group data by month (regex extract characters in date)
# mean temperature
# sum precipitation

# join on 'data' in sql, adding all 3 regions
```

```python
catalunya_meteo['date'] = catalunya_meteo['date'].str.extract('(.{7})')
catalunya_meteo
```

```python
# mean temperature, grouped by month

catalunya_temp = catalunya_meteo.groupby(by="date").temp_catalunya.mean()
catalunya_temp
```

```python
# sum precipitation, grouped by month

catalunya_precip = catalunya_meteo.groupby(by='date').precip_catalunya.sum()
catalunya_precip
```

```python
catalunya = pd.DataFrame([catalunya_temp, catalunya_precip]).T
catalunya
```

```python
# save clean dataset to catalunya_clean.csv

catalunya.to_csv("../data/catalunya_clean.csv", index=False)
```
