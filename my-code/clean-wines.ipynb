```python
# import libraries

import pandas as pd
import numpy as np
```

```python
# import dataset

wines = pd.read_csv('../data/ESPwines_dataset.csv')
```

```python
# explore the content of the file

wines.head(5)
```

```python
# see all columns - all are non-null, great :)

wines.info()
```

```python
# check for duplicated rows

wines.duplicated().sum()
```

```python
# drop duplicate rows

wines.drop_duplicates(inplace=True)
```

```python
wines
```

```python
wines.info()
```

```python
# the index seems to be off (2018 entries vs. 0 to 2024 index values), so it needs to be reset

wines = wines.reset_index()
wines.info()
```

```python
wines
```

```python
# rename regions as needed for this analysis: Rioja, Rib. del Duero, Catalunya, Other

catalunya = ['Cariñena', 'Priorato', 'Montsant', 
            'Cataluña', 'Tarragona', 'Terra Alta']
```

```python
wines['Selected_region'] = wines['Region'].apply(lambda x: 'Catalunya Sul' if x in catalunya 
                                                 else 'Rioja' if x == 'Rioja' 
                                                 else 'Ribera del Duero' if x == 'Ribera del Duero'
                                                 else 'other')
```

```python
wines
```

```python
# save clean dataset to wines_clean.csv

wines.to_csv("../data/wines_clean.csv", index=False)
```

```python
# test if the file was saved correctly, by loading it

wines_clean = pd.read_csv("../data/wines_clean.csv")
wines_clean
```
